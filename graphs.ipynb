```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

#### Quickly created tables from our external excel sheet to make figures for final report

```python
w2v_data = pd.DataFrame({'F1-Score': [.748, .732, .729, .728, .762, .716, .684, .741, .731, .782],
                         'Classifier': ['Logistic-Binary', 'Logistic-Multiclass', 'SVM', 'Random Forest', 'Gradient Booting Tree',
                                        'Logistic-Binary', 'Logistic-Multiclass', 'SVM', 'Random Forest', 'Gradient Booting Tree'],
                         'Word2Vec Embedding': ['GAD', 'GAD', 'GAD', 'GAD', 'GAD',
                                                'Pubmed', 'Pubmed', 'Pubmed', 'Pubmed', 'Pubmed']})
```

```python
sns.set(style='whitegrid', rc={'figure.figsize':(6,4)})
sns.despine()
ax = sns.barplot(x='Word2Vec Embedding', y='F1-Score', hue='Classifier', data=w2v_data)
ax.set_title('F1-score for classifiers using different Word2Vec embeddings')
ax.set_ylim(0.6, 0.8)
plt.legend(bbox_to_anchor=(.5, -.2), loc=1, borderaxespad=0)
```

```python
features_data = pd.DataFrame({'F1-score': [.75, .822, .748, .819, .583, .704, .657, .749],
                              'Features': ['Rule', 'Syntactic', 'Word2Vec', 'All',
                                           'Rule', 'Syntactic', 'Word2Vec', 'All'],
                              'Labels': ['Y+N/F', 'Y+N/F', 'Y+N/F', 'Y+N/F',
                                         'Y/N/F', 'Y/N/F', 'Y/N/F', 'Y/N/F']})
```

```python
sns.set(style='whitegrid', rc={'figure.figsize':(6,4)})
sns.despine()
ax = sns.barplot(x='Labels', y='F1-score', hue='Features', data=features_data)
ax.set_title('F1-score for binary fit logistic regression\n across different label sets and features')
ax.set_ylim(0, 1)
plt.legend(bbox_to_anchor=(.5, -.2), loc=1, borderaxespad=0)
```
