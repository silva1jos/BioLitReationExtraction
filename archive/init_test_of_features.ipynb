```python
def split(a, n):
    k, m = divmod(len(a), n)
    return (a[i * k + min(i, m):(i + 1) * k + min(i + 1, m)] for i in range(n))
```

```python
train_splits = split(train_features, 5)
train_labels_splits = split(train_labels, 5)
clfs = []
for data, labels in zip(train_splits, train_labels_splits):
    eclf = LogisticRegression(solver='lbfgs', multi_class='auto', max_iter=1000).fit(data, labels)
    clfs.append(eclf)
preds = [c.predict(test_features) for c in clfs]
pred = [max(set(x), key=x.count) for x in zip(*preds)]
sum([p == a for p, a in zip(pred,test_labels)]) / len(test_labels)
```
